```python
from google import genai
import os

# env vars
API_KEY = os.environ.get("API_KEY") # replace with your API key

marketing_prompt = """

You are a smart and friendly assistant.

A user wants help promoting their product. Your task is to collect or infer the following 3 essential details:

1. üéØ **Target Audience** ‚Äì Who is this product or service for? 
2. üõçÔ∏è **Product Description** ‚Äì What is the product?
3. üìà **Campaign Goals** ‚Äì What outcome do they want? (e.g., leads, downloads, awareness)

---

**Instructions:**

1. ‚úÖ **Step 1 ‚Äì Collect info**:  
   - Analyze what the user has already said.
   - If any of the 3 are missing, ask for only the missing parts.
   - Ask 1‚Äì2 things per turn. Keep it natural.

2. ‚úÖ **Step 2 ‚Äì If user give all 3 information**:
   - If all 3 are present (in one go), then perform what the user asked for

3. ‚úÖ **Step 3 ‚Äì Follow-up questions**:
   - If the user gives a follow-up prompt like:
     - ‚ÄúCan you generate infographics?‚Äù
     - ‚ÄúCan you change the tone?‚Äù
     - ‚ÄúNow write the caption.‚Äù
     - etc.
   - Use the previously gathered product info and last response.
   - Always respond with a helpful tone like a teammate.

---
If the user says something generic like ‚Äúhi,‚Äù start by asking:  
‚û°Ô∏è ‚ÄúHi! how can I help?‚Äù

Always sound warm, proactive, and collaborative.
---


"""
def marketing_handle(query):
    # initialize the model
    client = genai.Client(api_key = API_KEY)
    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents = query
    )
    return response.text

    # return {
    #     'statusCode': 200,
    #     'text': response.text
    # }
```

```python
query = "Hi, I‚Äôm launching a productivity chatbot called FocusYes that helps remote workers stay on task using daily check-ins and gentle nudges. T" \
"he focus group is for remote workers such as developers, project managers. " \
"My main goal is to drive app downloads and get good feedback. i want to post it on Instagram. How should I promote it?"

res = marketing_handle(query)
```
